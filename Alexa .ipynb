```python
import speech_recognition as sr
import pyttsx3
import pywhatkit
import datetime
import wikipedia
import pyjokes
import webbrowser as web
```

```python
listener = sr.Recognizer()#Voice recognizer
engine = pyttsx3.init()
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[1].id)#Femal Voice
#path = "C:\ProgramData\Microsoft\Windows\Start Menu\Programs/chrome.exe %s"

def talk(text):
    engine.say(text)
    engine.runAndWait()
    
def talk_command():
    
    try:
        with sr.Microphone() as source:#for microphone
            listener.adjust_for_ambient_noise(source)#removing noises
            print('listening...')
            voice = listener.listen(source)
            command = listener.recognize_google(voice)#Google API
            command = command.lower()
            if 'alexa' in command:
                command = command.replace('alexa', '')
                print(command)
            
    except:
        print("MicroPhone is not working")
        
    return command

def run_alexa():
    command = talk_command()
    print(command)
    if "play" in command:
        song = command.replace('play', '')
        talk('playing ' + song)
        pywhatkit.playonyt(song)#Play on youtube
    
    elif 'time' in command:
        time = datetime.datetime.now().strftime('%I:%M %p')
        talk('Current time is ' + time)
    
    elif 'who the heck is' in command:
        person = command.replace('who the heck is', '')
        info = wikipedia.summary(person, 1)
        print(info)
        talk(info)
    
    elif 'date' in command:
        talk('sorry, I have a headache')
    
    elif 'are you single' in command:
        talk('I am in a relationship with wifi')
    
    elif 'joke' in command:
        talk(pyjokes.get_joke())
         
    else:
        talk('Please say the command again.')

#while True:        
    #run_alexa()

if __name__ == "__main__":
    run_alexa()
```

```python
import speech_recognition as sr
import webbrowser as web

def main():

    path = "C:\ProgramData\Microsoft\Windows\Start Menu\Programs\chrome.exe %s"
    r = sr.Recognizer()

    with sr.Microphone() as source:
        r.adjust_for_ambient_noise(source)

        print("Please say something ")

        audio = r.listen(source)

        print("Reconizing Now ... ")



        try:
            dest = r.recognize_google(audio)
            print("You have said : " + dest)

            web.get(path).open(dest)

        except Exception as e:
            print("Error : " + str(e))


if __name__ == "__main__":
    main()
```
